```python
file_path ='/Volumes/workspace/default/atualizd_desmat/Desmatamento no brasil1 .csv'



df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("delimiter", ",") \
    .option("encoding", "ISO-8859-1") \
    .load(file_path)
 
display(df)
```

```python
df_filtrado = df.filter((df.referencia >= 2018) & (df.referencia <= 2022))
display(df_filtrado)
```

```python
df_pandas = df.toPandas()
```

```python
import matplotlib.pyplot as plt
import numpy as np

df_pandas = df.toPandas()

df_pandas["referencia"] = df_pandas["referencia"].astype(int)
df_pandas = df_pandas.sort_values("referencia")

estados_top9 = ["para", "mato_grosso", "rondonia", "amazonas", "acre", "amapa", "roraima", "tocantins","maranhao"]

plt.figure(figsize=(16, 8))

anos = df_pandas["referencia"].values
x = np.arange(len(anos))
largura = 0.10

for i, estado in enumerate(estados_top9):
    plt.bar(x + i * largura,
            df_pandas[estado],
            width=largura,
            label=estado.capitalize())

plt.xticks(x + largura * len(estados_top9) / 2, anos, rotation=45)

plt.title("Desmatamento  4 Estados que Mais Desmataram (20102022)", fontsize=16)
plt.xlabel("Ano", fontsize=13)
plt.ylabel("rea Desmatada (km)", fontsize=13)
plt.legend(title="Estados")
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
